```python
class MultiLayerNN:
    def __init__(self,fonction_Activation,nb_epochs,learning_rate,nbNCachee):
        self.hidden_bias=np.zeros((1,nbNCachee))
        self.output_bias=np.array([[0.25]])
        self.hidden_weights=np.random.rand(2,nbNCachee)
        self.output_weights=np.random.rand(nbNCachee,1)
        self.fonction_Activation=fonction_Activation
        self.nb_epochs=nb_epochs
        self.learning_rate=learning_rate
        self.nbNCachee=nbNCachee
      
    def sigmoid(x):
      return 1/(1 + np.exp(-x))

    def sigmoid_derivative(x):
      return x * (1 - x)
    
    def loss(yp,y):
      return (1/2)*np.square(yp-y)
        
    def fit(self,X,y):
        for _ in range(self.nb_epochs):
          #Learning process
          #Forward Pass
          nb_couches = 2
          hidden_layer_sum = np.dot(X,self.hidden_weights) + self.hidden_bias
          hidden_layer_out = sigmoid(hidden_layer_sum)
          output_layer_sum = np.dot(hidden_layer_out,self.output_weights) + self.output_bias
          predicted_y = sigmoid(output_layer_sum)

          #Backward pass
          delta=(predicted_y-y)*sigmoid_derivative(predicted_y)
          #hidden layer
          dloss_dw_h= X.T.dot(delta.dot(self.output_weights.T)*sigmoid_derivative(hidden_layer_out))
    
          dloss_dbias_h=(delta.dot(self.output_weights.T)*sigmoid_derivative(hidden_layer_out))
          #output layer
          dloss_dw_out=hidden_layer_out.T.dot(delta)
          dloss_dbias_out=delta


          #update
          self.hidden_weights-=dloss_dw_h*self.learning_rate
          self.hidden_bias=np.sum(-(dloss_dbias_h*self.learning_rate),axis=0,keepdims=True)
          self.output_weights-=dloss_dw_out*self.learning_rate
          self.output_bias-=np.sum(dloss_dbias_out,axis=0,keepdims=True)*self.learning_rate
          
        print("predicted : ",predicted_y)

#---------------------------------------
print("results------------------------------------")
print("updated hidden weights: ")
print(hidden_weights)
print("updated hidden biases: ")
#print(hidden_bias)
print("updated output weights: ")
#print(output_weights)
print("updated output biases: ")
#print(output_bias)
```
